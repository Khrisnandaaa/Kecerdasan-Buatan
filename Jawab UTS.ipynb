```python

# Data

company_detail_list = [
      {
          'name': 'Company 1',
          'domain': 'Retail',
          'country': 'United States'
      },
      {
          'name': 'Company 2',
          'domain': 'Technology',
          'country': 'United Kingdom'
      },
      {
          'name': 'Company 3',
          'domain': 'Healthcare',
          'country': 'United States'
      }
  ]

employee_detail_list = [
     {
          'name': 'EMP-0001',
          'first_name': 'John',
          'last_name': 'Doe',
          'full_name': 'John Doe',
          'company': 'Company 1',
          'nationality': 'Australia'
      },
      {
          'name': 'EMP-0002',
          'first_name': 'Tom',
          'last_name': 'Smith',
          'full_name': 'Tom Smith',
          'company': 'Company 2',
          'nationality': 'United States'
      },
      {
          'name': 'EMP-0003',
          'first_name': 'Andrew',
          'last_name': 'Sebastian',
          'full_name': 'Andrew Sebastian',
          'company': 'Company 3',
          'nationality': 'United States'
      },
      {
          'name': 'EMP-0005',
          'first_name': 'Ying Han',
          'last_name': 'Tan',
          'full_name': 'Ying Han Tan',
          'company': 'Company 1',
          'nationality': 'Australia'
      },
      {
          'name': 'EMP-0015',
          'first_name': 'Kenneth',
          'last_name': 'Ng',
          'full_name': 'Kenneth Ng',
          'company': 'Company 3',
          'nationality': 'United States'
      },
      {
          'name': 'EMP-0018',
          'first_name': 'Rubby',
          'last_name': 'Lee',
          'full_name': 'Rubby Lee',
          'company': 'Company 2',
          'nationality': 'Hong Kong'
      },
     {
          'name': 'EMP-0017',
          'first_name': 'Robert',
          'last_name': 'White',
          'full_name': 'Robert White',
          'company': 'Company 1',
          'nationality': 'United Kingdom'
      }
  ]


Task 1

Please get the list of all Companies and sort by **Company Name** in reverse order.

[
    {
        "name": "Company 3"
    },
    {
        "name": "Company 2"
    },
    {
        "name": "Company 1"
    }
]


Task 2

Please print all Domain value in every company (use print statement).

company_detail_list = [
    {
        'name': 'Company 1',
        'domain': 'Retail',
        'country': 'United States'
    },
    {
        'name': 'Company 2',
        'domain': 'Technology',
        'country': 'United Kingdom'
    },
    {
        'name': 'Company 3',
        'domain': 'Healthcare',
        'country': 'United States'
    }
]

for company in company_detail_list:
    print(f"{company['name']}: {company['domain']} ({company['country']})")


Task 3

List all Employees work By company domain.

company_detail_list = [
    {
        'name': 'Company 1',
        'domain': 'Retail',
        'country': 'United States'
    },
    {
        'name': 'Company 2',
        'domain': 'Technology',
        'country': 'United Kingdom'
    },
    {
        'name': 'Company 3',
        'domain': 'Healthcare',
        'country': 'United States'
    }
]

employee_detail_list = [
    {
        'name': 'EMP-0001',
        'full_name': 'John Doe',
        'company': 'Company 1',
    },
    {
        'name': 'EMP-0002',
        'full_name': 'Tom Smith',
        'company': 'Company 2',
    },
    {
        'name': 'EMP-0003',
        'full_name': 'Andrew Sebastian',
        'company': 'Company 3',
    },
    {
        'name': 'EMP-0005',
        'full_name': 'Ying Han Tan',
        'company': 'Company 1',
    },
    {
        'name': 'EMP-0015',
        'full_name': 'Kenneth Ng',
        'company': 'Company 3',
    },
    {
        'name': 'EMP-0018',
        'full_name': 'Rubby Lee',
        'company': 'Company 2',
    },
    {
        'name': 'EMP-0017',
        'full_name': 'Robert White',
        'company': 'Company 1',
    }
]

# Create a mapping of company names to their domains
company_domain_map = {company['name']: company['domain'] for company in company_detail_list}

# Create a dictionary to group employees by domain
employees_by_domain = {}
for employee in employee_detail_list:
    domain = company_domain_map[employee['company']]
    if domain not in employees_by_domain:
        employees_by_domain[domain] = []
    employees_by_domain[domain].append(employee['full_name'])

# Print the result
print(employees_by_domain)


Task 4

Create a function that return employees with it's company country.

company_detail_list = [
    {
        'name': 'Company 1',
        'domain': 'Retail',
        'country': 'United States'
    },
    {
        'name': 'Company 2',
        'domain': 'Technology',
        'country': 'United Kingdom'
    },
    {
        'name': 'Company 3',
        'domain': 'Healthcare',
        'country': 'United States'
    }
]

employee_detail_list = [
    {
        'name': 'EMP-0001',
        'full_name': 'John Doe',
        'company': 'Company 1',
    },
    {
        'name': 'EMP-0002',
        'full_name': 'Tom Smith',
        'company': 'Company 2',
    },
    {
        'name': 'EMP-0003',
        'full_name': 'Andrew Sebastian',
        'company': 'Company 3',
    },
    {
        'name': 'EMP-0005',
        'full_name': 'Ying Han Tan',
        'company': 'Company 1',
    },
    {
        'name': 'EMP-0015',
        'full_name': 'Kenneth Ng',
        'company': 'Company 3',
    },
    {
        'name': 'EMP-0018',
        'full_name': 'Rubby Lee',
        'company': 'Company 2',
    },
    {
        'name': 'EMP-0017',
        'full_name': 'Robert White',
        'company': 'Company 1',
    }
]

def get_employees_with_country():
    # Create a mapping of company names to their country
    company_country_map = {company['name']: company['country'] for company in company_detail_list}
    
    # Generate the list with full name, company, and country
    employee_list = []
    for employee in employee_detail_list:
        company = employee['company']
        employee_list.append({
            'full_name': employee['full_name'],
            'company': company,
            'country': company_country_map[company]
        })
    
    return employee_list

# Call the function and print the result
result = get_employees_with_country()
print(result)


Task 5

Create a function that return companies with a number of employees nationality that are working in the company.

company_detail_list = [
    {
        'name': 'Company 1',
        'domain': 'Retail',
        'country': 'United States'
    },
    {
        'name': 'Company 2',
        'domain': 'Technology',
        'country': 'United Kingdom'
    },
    {
        'name': 'Company 3',
        'domain': 'Healthcare',
        'country': 'United States'
    }
]

employee_detail_list = [
    {
        'name': 'EMP-0001',
        'full_name': 'John Doe',
        'company': 'Company 1',
        'nationality': 'Australia'
    },
    {
        'name': 'EMP-0002',
        'full_name': 'Tom Smith',
        'company': 'Company 2',
        'nationality': 'United States'
    },
    {
        'name': 'EMP-0003',
        'full_name': 'Andrew Sebastian',
        'company': 'Company 3',
        'nationality': 'United States'
    },
    {
        'name': 'EMP-0005',
        'full_name': 'Ying Han Tan',
        'company': 'Company 1',
        'nationality': 'Australia'
    },
    {
        'name': 'EMP-0015',
        'full_name': 'Kenneth Ng',
        'company': 'Company 3',
        'nationality': 'United States'
    },
    {
        'name': 'EMP-0018',
        'full_name': 'Rubby Lee',
        'company': 'Company 2',
        'nationality': 'Hong Kong'
    },
    {
        'name': 'EMP-0017',
        'full_name': 'Robert White',
        'company': 'Company 1',
        'nationality': 'United Kingdom'
    }
]

def get_company_nationality_counts():
    # Initialize a dictionary to store the result
    company_nationality_count = {company['name']: {} for company in company_detail_list}
    
    # Count employees' nationalities for each company
    for employee in employee_detail_list:
        company = employee['company']
        nationality = employee['nationality']
        
        # Update the count for each nationality
        if nationality in company_nationality_count[company]:
            company_nationality_count[company][nationality] += 1
        else:
            company_nationality_count[company][nationality] = 1
    
    return company_nationality_count

# Call the function and print the result
result = get_company_nationality_counts()
print(result)
```
